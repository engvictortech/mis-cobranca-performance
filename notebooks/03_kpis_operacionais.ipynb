### 🎯 Objetivo

Este notebook tem como objetivo transformar dados operacionais de cobrança em inteligência estratégica.

A análise contempla:

- Estrutura da carteira de inadimplência
- Perfil de atraso dos clientes
- Volume e distribuição de acionamentos
- Efetividade de contato
- Conversão para promessa de pagamento
- Indicadores preliminares de recuperação

---

### 🧠 Abordagem

A análise combina visão:

- Técnica (estatística descritiva e agregações)
- Operacional (produtividade e funil de cobrança)
- Executiva (indicadores estratégicos para tomada de decisão)

Este material servirá de base para construção dos KPIs e dashboards executivos.


```python
# ==========================================================
# 📦 IMPORTAÇÃO DE BIBLIOTECAS
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ==========================================================
# 📂 LEITURA DA BASE CONSOLIDADA
# ==========================================================

base = pd.read_csv("../data/processed/base_consolidada.csv")

print("Dimensão da base:", base.shape)
```

```python
# ==========================================================
# 🔎 VALIDAÇÃO DA BASE
# ==========================================================

base.info()
base.describe()
```

```python
# ==========================================================
# ⚠️ VERIFICAÇÃO DE VALORES NULOS
# ==========================================================

base.isnull().sum()
```

```python
# ==========================================================
# 📊 KPI - TICKET MÉDIO DE DÍVIDA
# ==========================================================

ticket_medio = base["valor_divida"].mean()
print("Ticket Médio:", round(ticket_medio, 2))

```

```python
# ==========================================================
# 📊 KPI - TAXA GERAL DE RECUPERAÇÃO
# ==========================================================

taxa_recuperacao = base["cliente_recuperado"].mean()
print("Taxa Geral de Recuperação:", round(taxa_recuperacao * 100, 2), "%")
```

```python
# ==========================================================
# 📊 KPI - MÉDIA PERCENTUAL RECUPERADO
# ==========================================================

media_recuperado = base["percentual_recuperado"].mean()
print("Percentual Médio Recuperado:", round(media_recuperado * 100, 2), "%")
```

```python
# ==========================================================
# 📊 KPI - MÉDIA DE ACIONAMENTOS POR CLIENTE
# ==========================================================

media_acionamentos = base["total_acionamentos"].mean()
print("Média de Acionamentos:", round(media_acionamentos, 2))
```

```python
# ==========================================================
# 💾 EXPORTAÇÃO FINAL
# ==========================================================

base.to_csv("../data/processed/base_consolidada_final.csv", index=False)
```
